```python
import os
```

```python
def rename_files(path, start_count):
    image_list = []
    file_count = start_count
    for file in os.listdir(path):
        filename_source = path + file
        file_ext = file.split('.')[1]
        filename_renamed = path + f'{file_count}.{file_ext}'
        os.rename(filename_source, filename_renamed)
        file_count += 1
        print(filename_renamed)
```
